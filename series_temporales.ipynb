```python
import matplotlib.pyplot as plt
import numpy as np

# ======== Generar datos ========
np.random.seed(1)
periodo = np.arange(1, 26)

# Crear una tendencia lineal + componente estacional + ruido leve
tendencia = 9 + 1.0 * periodo
estacionalidad = 4 * np.sin(periodo * np.pi / 6)
ruido = np.random.normal(0, 1, len(periodo))
valor = tendencia + estacionalidad + ruido

# ======== Graficar ========
plt.figure(figsize=(8,5))
plt.plot(periodo, valor, 'o-', color='steelblue', linewidth=2, markersize=6)
plt.title('LÃ­neas: serie temporal con tendencia y estacionalidad', fontsize=12)
plt.xlabel('Periodo')
plt.ylabel('Valor')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
