```python
import numpy as np
import matplotlib.pyplot as plt

cats = ['Norte', 'Sur', 'Este', 'Oeste']
conteos = np.array([120, 90, 150, 110])

plt.figure(figsize=(8, 6))
plt.bar(cats, conteos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Categorías')
plt.ylabel('Conteos')
plt.title('Conteos por Categoría')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

El gráfico de barras es la mejor opción para tus datos porque:
Representa comparaciones entre categorías de forma directa, clara y cuantitativamente precisa, maximizando la comprensión visual.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 120
x = np.linspace(0, 100, n)
y = 2.0 * x + np.random.normal(0, 25, n)  # pendiente ~2 + ruido

plt.figure(figsize=(10, 7))
plt.scatter(x, y, label='Datos con Ruido', alpha=0.7, color='purple')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Dispersión de Datos con Pendiente y Ruido')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

El gráfico de dispersión es la mejor opción para este tipo de datos porque:
Permite visualizar de manera directa y clara la relación entre dos variables numéricas, identificar la tendencia general, observar la dispersión causada por el ruido y analizar posibles correlaciones.

```python
import numpy as np
import matplotlib.pyplot as plt

muestra = np.random.normal(loc=50, scale=10, size=1000) + np.random.exponential(scale=2, size=1000)*0.5

plt.figure(figsize=(10, 6))
plt.hist(muestra, bins=50, density=True, alpha=0.7, color='teal', label='Distribución de la Muestra')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.title('Histograma de la Muestra')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

El histograma es la mejor opción en este caso porque:
Permite visualizar la distribución de una variable continua, mostrando su forma, dispersión y tendencia central, lo cual es esencial para entender el comportamiento de los datos generados aleatoriamente.

```python
import matplotlib.pyplot as plt
import numpy as np


t = np.arange(1, 25)
tendencia = 5 + 1.2 * t
estacional = 4 * np.sin(2 * np.pi * t / 12)
ruido = np.random.normal(0, 2, size=t.shape)
serie = tendencia + estacional + ruido


fig, ax = plt.subplots(figsize=(10, 6))


burbujas = ax.scatter(t, serie, s=np.abs(estacional)*10, alpha=0.6, c=estacional, cmap='viridis', edgecolors='w', linewidth=0.5)


ax.set_xlabel("Tiempo ($t$)")
ax.set_ylabel("Serie temporal")
ax.set_title("Gráfica de burbujas con componentes de tendencia, estacionalidad y ruido")


leyenda_tamano = plt.legend(*burbujas.legend_elements(prop="sizes", num=4),
                            title="Componente estacional", loc="upper left")
ax.add_artist(leyenda_tamano)
plt.colorbar(burbujas, label="Valor de la componente estacional")

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

El gráfico de burbujas aplicado a una serie temporal es la mejor opción en este caso porque:
Permite representar simultáneamente la tendencia, la estacionalidad y el ruido de una serie, mostrando múltiples dimensiones de información (tiempo, valor, magnitud estacional y fase) en una sola visualización intuitiva y visualmente rica.

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 500
T = 2.0
t = np.linspace(0, T, int(fs*T), endpoint=False)
f1, f2 = 50, 120
signal = 1.0*np.sin(2*np.pi*f1*t) + 0.6*np.sin(2*np.pi*f2*t) + 0.3*np.random.normal(0,1,len(t))

Y = np.fft.rfft(signal)
freqs = np.fft.rfftfreq(len(t), d=1/fs)
amplitud = np.abs(Y) / len(t)

fig, ax2 = plt.subplots(figsize=(12, 5))

# Plot the frequency spectrum (amplitude spectrum)
ax2.stem(freqs, amplitud, linefmt='red', markerfmt='o', basefmt=' ', label='Espectro de Amplitud')
ax2.set_xlabel('Frecuencia (Hz)')
ax2.set_ylabel('Amplitud Normalizada')
ax2.set_title('Espectro de Frecuencia de la Señal')
ax2.set_xlim(0, fs/2) # Limit x-axis to Nyquist frequency
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Este tipo de gráfico es ideal para análisis de señales porque permite:
Identificar frecuencias dominantes de manera rápida y precisa.
Diferenciar componentes de señal y ruido.
Representar la información de forma visualmente clara y técnica, perfecta para ingeniería, física o procesamiento de señales.
